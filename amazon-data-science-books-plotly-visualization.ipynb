# <font color='blue'> Import libraries </font> <a class="anchor" id="Importlibrary"></a>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# <font color='blue'> Import data </font> <a class="anchor" id="a"></a>

```python
data = pd.read_csv('/kaggle/input/amazon-data-science-books/final_book_dataset_kaggle.csv')
data.head()
```

# <font color='blue'> Dataset description </font> <a class="anchor" id="b"></a>

```python
data.describe()
```

```python
data.isna().sum()
```

# <font color='blue'> 'Parallel_coordinates' plot</font> <a class="anchor" id="c"></a>

```python
fig = px.parallel_coordinates(data, color="price", labels=["star5", "star4",
                  "star3", "star2", "star1", "pages",
                  "n_reviews"],color_continuous_scale=['rgb(10,6,135)','rgb(181,228,255)',
                     'rgb(215,73,12)','rgb(165,13,13)','rgb(119,50,61)'],
                    color_continuous_midpoint=137)
fig.show()
```

# <font color='blue'> Plotly scatterplots </font> <a class="anchor" id="d"></a>

```python
px.scatter(data, x="price", y="pages",color='avg_reviews',color_continuous_scale='RdBu_r'
          ,size='n_reviews')

#fig.show()
```

```python
a=(data['n_reviews']+110).tolist()
fig = px.scatter(data, x="price", y="pages", size=a,facet_col="avg_reviews",color='star5',color_continuous_scale='RdBu_r'
                )

'''
import plotly.graph_objects as go
size=data['n_reviews'].tolist()
fig = go.Figure(data=go.Scatter(x=data['price'], y=data["pages"], size='n_reviews',facet_col="avg_reviews",
    mode='markers',
    marker_size=dict(
        size=size,
        sizemode='area',
        sizeref=2.*max(size)/(40**2)
    )
                                ))
                                '''
fig.update_layout(
    width = 11100,
    height = 640,
    title = "Price, Number of Pages, Average Reviews and Percentage of 5 stars per Average Review"
)


fig.show()
```

# <font color='blue'> Plotly barplots </font> <a class="anchor" id="e"></a>

```python
datalargest=data['n_reviews'].nlargest(15).astype(int).sort_index()
fifteen_most_reviewed_books=data[data['n_reviews'] >= datalargest.min()]
```

```python
fig = px.bar(fifteen_most_reviewed_books, x='title', y='avg_reviews' ,log_y=True)
fig.update_layout(
    width = 800,
    height = 1640,
    title = "Fifteen Most Reviewed Books: Average Reviews"
)
fig.show()
```

```python
fig = px.bar(fifteen_most_reviewed_books, x='title', y=['star5','star4','star3','star2','star1'],log_y=True)
fig.update_layout(
    width = 800,
    height = 1640,
    title = "Fifteen Most Reviewed Books: Number of Stars Percentage"
)
fig.show()
```

```python
data_price_largest=data['price'].nlargest(15).astype(int).sort_index()
fifteen_most_expensive_books=data[data['price'] >= data_price_largest.min()]
```

```python
fig = px.bar(fifteen_most_expensive_books, x='title', y='avg_reviews' ,log_y=True)
fig.update_layout(
    width = 800,
    height = 1640,
    title = "Fifteen Most Expensive Books: Average Reviews"
)
fig.show()
```

```python
fig =px.bar(fifteen_most_expensive_books, x='title', y=['star5','star4','star3','star2','star1'])
fig.update_layout(
    width = 800,
    height = 1340,
    title = "Fifteen Most Expensive Books: Number of Stars Percentage"
)
fig.show()
```
